```python
from itertools import product

# Function to calculate the next secret number
def next_secret_number(secret):
    # Step 1: Multiply by 64, mix, and prune
    secret = (secret ^ (secret * 64)) % 16777216
    # Step 2: Divide by 32 (rounded down), mix, and prune
    secret = (secret ^ (secret // 32)) % 16777216
    # Step 3: Multiply by 2048, mix, and prune
    secret = (secret ^ (secret * 2048)) % 16777216
    return secret

# Function to generate prices from secret numbers
def generate_prices(initial_secret, count):
    secret = initial_secret
    prices = []
    for _ in range(count):
        prices.append(secret % 10)  # Price is the ones digit of the secret number
        secret = next_secret_number(secret)
    return prices

# Function to calculate price changes
def calculate_price_changes(prices):
    return [prices[i] - prices[i - 1] for i in range(1, len(prices))]

# Function to find the first occurrence of a sequence in a list
def find_sequence(sequence, changes):
    for i in range(len(changes) - len(sequence) + 1):
        if changes[i:i + len(sequence)] == sequence:
            return i + len(sequence)  # Return the index after the sequence
    return None

# Main function
if __name__ == "__main__":
    # Read input file containing initial secret numbers
    with open("/content/drive/MyDrive/Personal Project/Advent of Code/2024/Day_22/input_22_12_2024.txt", "r") as file:
        initial_secrets = [int(line.strip()) for line in file.readlines()]

    # Generate all possible sequences of four price changes
    possible_sequences = list(product(range(-9, 10), repeat=4))

    # Simulate buyers and calculate bananas for each sequence
    max_bananas = 0
    best_sequence = None

    for sequence in possible_sequences:
        total_bananas = 0
        for secret in initial_secrets:
            prices = generate_prices(secret, 2001)  # Generate 2001 prices to calculate 2000 changes
            changes = calculate_price_changes(prices)
            index = find_sequence(sequence, changes)
            if index is not None:
                total_bananas += prices[index]  # Add the price where the sequence occurs

        if total_bananas > max_bananas:
            max_bananas = total_bananas
            best_sequence = sequence

    # Print the result
    print(f"Best sequence: {best_sequence}")
    print(f"Maximum bananas: {max_bananas}")
```
