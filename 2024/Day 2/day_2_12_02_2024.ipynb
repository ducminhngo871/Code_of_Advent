```python
import pandas as pd

# Load the Excel file
file_path = "/content/drive/MyDrive/Personal Project/Advent of Code/2024/Day_2/day_2_12_02_2024.xlsx"  # Replace with your Excel file path
sheet_name = "Sheet1"         # Replace with your sheet name if needed
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Convert the data to numeric and drop NaN values
data = data.apply(pd.to_numeric, errors='coerce').dropna(how='all')

# Convert each row into a list of numbers (ignoring NaN values)
reports = data.apply(lambda row: row.dropna().tolist(), axis=1).tolist()

# Define the function to check if a report is safe
def is_safe(report):
    # Check if the list is increasing or decreasing
    is_increasing = all(1 <= report[i + 1] - report[i] <= 3 for i in range(len(report) - 1))
    is_decreasing = all(-3 <= report[i + 1] - report[i] <= -1 for i in range(len(report) - 1))
    return is_increasing or is_decreasing

# Define the function to check safety with the Problem Dampener
def is_safe_with_dampener(report):
    # If the report is already safe, return True
    if is_safe(report):
        return True

    # Try removing each level and check if the modified report is safe
    for i in range(len(report)):
        modified_report = report[:i] + report[i + 1:]  # Remove the i-th element
        if is_safe(modified_report):
            return True

    # If no modification makes the report safe, return False
    return False

# Count the number of safe reports
safe_reports = sum(is_safe_with_dampener(report) for report in reports)
print("Number of Safe Reports (with Problem Dampener):", safe_reports)
```
