```python
# Function to parse the input and separate locks and keys
def parse_input(file_path):
    with open(file_path, 'r') as f:
        data = f.read().strip().split('\n\n')  # Split into blocks of lock/key schematics

    locks, keys = [], []
    for block in data:
        rows = block.split('\n')
        if all(char == '#' for char in rows[0]):  # Identify locks by first row
            locks.append(block)
        elif all(char == '#' for char in rows[-1]):  # Identify keys by last row
            keys.append(block)

    return locks, keys

# Function to calculate the heights of locks and keys
def calculate_heights(schematic):
    rows = schematic.split('\n')
    heights = []
    for col in range(len(rows[0])):  # Iterate through columns
        height = sum(1 for row in rows[1:-1] if row[col] == '#')  # Exclude first and last rows
        heights.append(height)
    return heights

# Function to check compatibility between a lock and a key
def is_compatible(lock_heights, key_heights):
    for l, k in zip(lock_heights, key_heights):
        if l + k > 5:
            return False
    return True

# Main function to process the input and count compatible pairs
def count_compatible_pairs(file_path):
    locks, keys = parse_input(file_path)

    # Calculate heights for locks and keys
    lock_heights = [calculate_heights(lock) for lock in locks]
    key_heights = [calculate_heights(key) for key in keys]

    # Check compatibility
    compatible_pairs = 0
    for l_height in lock_heights:
        for k_height in key_heights:
            if is_compatible(l_height, k_height):
                compatible_pairs += 1

    return compatible_pairs

# Example usage
if __name__ == "__main__":
    file_path = "/content/drive/MyDrive/Personal Project/Advent of Code/2024/Day_25/input_25_12_2024.txt"  # Replace with your input file path
    result = count_compatible_pairs(file_path)
    print(f"Number of compatible lock/key pairs: {result}")
```
